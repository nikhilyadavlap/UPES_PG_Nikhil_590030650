# Question 5 – String Concatenation

## Aim
To learn how to concatenate (join) two strings using variables and display the full name in Python.

## Theory
String concatenation is the process of joining two or more strings into a single string. In Python, the `+` operator is used to combine strings. A space can be added between strings using `" "`.

## Program

```python

```

## Output

```

```

## Explanation

- `a` stores the first name.
- `b` stores the last name.
- The `+` operator joins both strings.
- `" "` adds a space between the first name and last name.
- `print()` displays the complete name.

## Key Concepts

- Variables
- Strings (`str`)
- String Concatenation
- `+` Operator
- `print()` Function

## Conclusion

This program demonstrates how to combine two string variables using the concatenation operator (`+`) to create and display a full name.

# Question 6 – Formatted String (f-String)

## Aim
To declare the first name, last name, and nickname, and display them in the format **FirstName (Nickname) LastName** using Python f-strings.

## Theory
An **f-string** (formatted string literal) is a modern way to format strings in Python. It is created by placing the letter `f` before a string. Variables enclosed in curly braces `{}` are automatically replaced with their values.

## Program

```python
first_name = "Nikhil"
last_name = "Yadav"
nickname = "Nicky"

print(f"{first_name} ({nickname}) {last_name}")
```

## Output

```
Nikhil (Nicky) Yadav
```

## Explanation

- `first_name` stores the first name.
- `last_name` stores the last name.
- `nickname` stores the nickname.
- The prefix `f` creates an f-string.
- Curly braces `{}` insert the values of variables into the string.
- `print()` displays the formatted output.

## Key Concepts

- Variables
- Strings (`str`)
- f-Strings
- String Formatting
- Curly Braces `{ }`
- `print()` Function

## Conclusion

This program demonstrates how to use Python f-strings to combine multiple variables into a single formatted string. F-strings make the code more readable, efficient, and easier to maintain.
