# Question 1 – Check Whether a Number is Divisible by 3 and 5

## Aim
To write a Python program that checks whether a given number is divisible by both 3 and 5.

## Algorithm

1. Start the program.
2. Assign a value to the variable `num`.
3. Check whether `num` is divisible by 3 using `%`.
4. Check whether `num` is divisible by 5 using `%`.
5. If both conditions are true, print that the number is divisible by both 3 and 5.
6. Otherwise, print that it is not divisible by both.
7. Stop the program.

## Program

```python

    print(num, "is divisible by both 3 and 5")
else:
    print(num, "is not divisible by both 3 and 5")
```

```

```

## Method Used

- **Selection (Decision Making)** using the `if-else` statement.
- **Modulus Operator (`%`)** to check divisibility.
- **Logical AND (`and`)** to combine two conditions.

## Result

The program successfully checks whether the given number is divisible by both **3** and **5** and displays the appropriate message.

# Question 5 – Check Whether a Quadratic Equation Has Real or Imaginary Roots

## Aim
To write a Python program to determine whether the roots of a quadratic equation are real or imaginary and display the roots.

## Algorithm

1. Start the program.
2. Import the `cmath` module.
3. Assign values to `a`, `b`, and `c`.
4. Calculate the discriminant:
   $begin:math:display$
   D \= b\^2 \- 4ac
   $end:math:display$
5. Find the two roots using the quadratic formula.
6. If the discriminant is greater than or equal to 0, print that the roots are real.
7. Otherwise, print that the roots are imaginary.
8. Display both roots.
9. Stop the program.

## Program

```python
import cmath

a = 1
b = 2
c = 5

d = b**2 - 4*a*c

root1 = (-b + cmath.sqrt(d)) / (2*a)
root2 = (-b - cmath.sqrt(d)) / (2*a)

if d >= 0:
    print("Roots are Real")
else:
    print("Roots are Imaginary")

print("Root 1 =", root1)
print("Root 2 =", root2)
```

## Output

```
Roots are Imaginary
Root 1 = (-1+2j)
Root 2 = (-1-2j)
```

## Method Used

- **Conditional Statement:** `if-else`
- **Mathematical Formula:** Discriminant (`b² - 4ac`)
- **Module:** `cmath`
- **Comparison Operator:** `>=`

## Result

The program successfully determines whether the roots of a quadratic equation are **real or imaginary** and displays both roots correctly.

## Question 5

Check whether a quadratic equation has real or imaginary roots and display the roots.

**Equation format:** `ax^2 + bx + c = 0`

````python
import cmath

a, b, c = 1, 2, 5

discriminant = b ** 2 - 4 * a * c
root1 = (-b + cmath.sqrt(discriminant)) / (2 * a)
root2 = (-b - cmath.sqrt(discriminant)) / (2 * a)

if discriminant >= 0:
    print("Roots are real")
else:
    print("Roots are imaginary")

print("Root 1:", root1)
print("Root 2:", root2)
````

# Question 7 – Display the Next Date in the Calendar

## Aim
To write a Python program to take a date and display the next date using the `datetime` module.

# Question 8 – Print the Grade Sheet of a Student

## Aim
To write a Python program to prepare a student's grade sheet by calculating the total marks, percentage, CGPA, and grade based on marks obtained in five subjects.

## Algorithm

1. Start the program.
2. Create a function to assign a grade based on the CGPA.
3. Enter student details (Name, Roll Number, SAP ID, Semester, Course).
4. Store the marks of five subjects.
5. Calculate the total marks.
6. Calculate the percentage using:

   **Percentage = (Total Marks / 500) × 100**

7. Calculate the CGPA using:

   **CGPA = Percentage / 10**

8. Find the grade using the grade function.
9. Display the student's details, subject-wise marks, percentage, CGPA, and grade.
10. Stop the program.

## Program

```python
def get_grade(cgpa):
    if 0.0 <= cgpa <= 3.4:
        return "F"
    elif 3.5 <= cgpa <= 5.0:
        return "C+"
    elif 5.1 <= cgpa <= 6.0:
        return "B"
    elif 6.1 <= cgpa <= 7.0:
        return "B+"
    elif 7.1 <= cgpa <= 8.0:
        return "A"
    elif 8.1 <= cgpa <= 9.0:
        return "A+"
    elif 9.1 <= cgpa <= 10.0:
        return "O (Outstanding)"
    else:
        return "Invalid CGPA"

name = "Nikhil Yadav"
roll_no = "500030650"
sap_id = "590030650"
semester = 1
course = "MCA"

marks = {
    "PDS": 70,
    "Python": 80,
    "Chemistry": 90,
    "English": 60,
    "Physics": 65
}

total = sum(marks.values())
percentage = (total / 500) * 100
cgpa = round(percentage / 10, 1)
grade = get_grade(cgpa)

print("Sample Gradesheet")
print("Name:", name)
print("Roll Number:", roll_no, "SAP ID:", sap_id)
print("Semester:", semester, "Course:", course)

print("\nSubject\t\tMarks")
for subject, score in marks.items():
    print(subject, "\t", score)

print("\nTotal Marks:", total)
print("Percentage: {:.2f}%".format(percentage))
print("CGPA:", cgpa)
print("Grade:", grade)
```

## Question 8

Print the grade sheet of a student from marks of five subjects.

- `Percentage = (total marks / 500) * 100`
- `CGPA = Percentage / 10`
- Grade is assigned using the provided CGPA ranges.

````python
def get_grade(cgpa):
    if 0.0 <= cgpa <= 3.4:
        return "F"
    if 3.5 <= cgpa <= 5.0:
        return "C+"
    if 5.1 <= cgpa <= 6.0:
        return "B"
    if 6.1 <= cgpa <= 7.0:
        return "B+"
    if 7.1 <= cgpa <= 8.0:
        return "A"
    if 8.1 <= cgpa <= 9.0:
        return "A+"
    if 9.1 <= cgpa <= 10.0:
        return "O (Outstanding)"
    return "Invalid CGPA"

name = "Nikhil Yadav"
roll_no = "500030650"
sap_id = "590030650"
semester = 1
course = "MCA"

marks = {
    "PDS": 70,
    "Python": 80,
    "Chemistry": 90,
    "English": 60,
    "Physics": 65,
}

total = sum(marks.values())
percentage = (total / (len(marks) * 100)) * 100
cgpa = round(percentage / 10, 1)
grade = get_grade(cgpa)

print("Sample Gradesheet")
print("Name:", name)
print("Roll Number:", roll_no, "SAPID:", sap_id)
print("Sem:", semester, "Course:", course)
print("Subject name: Marks")
for subject, score in marks.items():
    print(f"{subject}: {score}")
print(f"Percentage: {percentage:.2f}%")
print("CGPA:", cgpa)
print("Grade:", grade)
````
