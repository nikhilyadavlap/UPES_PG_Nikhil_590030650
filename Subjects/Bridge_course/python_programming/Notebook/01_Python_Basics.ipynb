## 5. Intensive: Input, Conversion, and Formatting

Run the cell, change the sample values, and explain why conversion is required before numeric calculation.

```python
name = "Asha"
age_text = "18"
marks_text = "86.5"

age = int(age_text)
marks = float(marks_text)

print(f"Student: {name}")
print(f"Age next year: {age + 1}")
print(f"Marks after bonus: {marks + 2.5:.2f}")
```

## 6. Intensive: Truthiness and Type Checks

```python
values = [0, 1, "", "UPES", [], [10, 20], None, True]

for value in values:
    print(repr(value), "=>", bool(value), "=>", type(value).__name__)
```
